```python
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans, AffinityPropagation
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
X, _ = make_blobs(n_samples=100,
                n_features= 2 
              , cluster_std=3.0,
              random_state=1)
```

```python
print(_)
```

```python
plt.scatter(X[:,0], X[:,1])
plt.show()
```

```python
model = KMeans(n_clusters= 3 , init='random', n_init=1, max_iter=10, random_state=1, algorithm="full")#클러스터의 개수, 뿌리는 방법, 
model.fit(X) 
```

```python
cluster_centers = model.cluster_centers_
labels = model.labels_
print(cluster_centers) #3개의 클러스터에서의 중심이 나온다. 
print(labels) #각 원소의 배정된 결과를 나타내준다. 
```

```python
plt.scatter(X[labels==0,0], X[labels==0 , 1], 
            marker= "^", facecolor= 'red')
plt.scatter(X[labels==1,0], X[labels==1 , 1], 
            marker= "s", facecolor= 'green')
plt.scatter(X[labels==2,0], X[labels==2 , 1], 
            marker= "d", facecolor= 'blue')
plt.scatter(cluster_centers[0][0], cluster_centers[0][1],
            marker= "^", facecolor= 'red', s=300, edgecolors="k")
plt.scatter(cluster_centers[1][0], cluster_centers[1][1],
            marker= "s", facecolor= 'green', s=300, edgecolors="k")
plt.scatter(cluster_centers[2][0], cluster_centers[2][1],
            marker= "d", facecolor= 'blue', s=300, edgecolors="k")

```

```python
def plot_kmeans(n):
    model = KMeans(n_clusters= 3 , init='random', n_init=1, max_iter=n, random_state=1, algorithm="full").fit(X) 
    cluster_centers = model.cluster_centers_
    labels = model.labels_
    plt.scatter(X[labels==0,0], X[labels==0 , 1], 
            marker= "^", facecolor= 'red')
    plt.scatter(X[labels==1,0], X[labels==1 , 1], 
            marker= "s", facecolor= 'green')
    plt.scatter(X[labels==2,0], X[labels==2 , 1], 
            marker= "d", facecolor= 'blue')
    plt.scatter(cluster_centers[0][0], cluster_centers[0][1],
            marker= "^", facecolor= 'red', s=300, edgecolors="k")
    plt.scatter(cluster_centers[1][0], cluster_centers[1][1],
            marker= "s", facecolor= 'green', s=300, edgecolors="k")
    plt.scatter(cluster_centers[2][0], cluster_centers[2][1],
            marker= "d", facecolor= 'blue', s=300, edgecolors="k")
    plt.show()

for n in range(10):
  plot_kmeans(n+1)
```

```python
for i in range(1,10):
  model = KMeans(n_clusters= 3 , init='random', n_init=1, max_iter=i, random_state=1, algorithm="full").fit(X) 
  cluster_centers = model.cluster_centers_
  labels = model.labels_
  plt.scatter(X[labels==0,0], X[labels==0 , 1], 
            marker= "^", facecolor= 'red')
  plt.scatter(X[labels==1,0], X[labels==1 , 1], 
            marker= "s", facecolor= 'green')
  plt.scatter(X[labels==2,0], X[labels==2 , 1], 
            marker= "d", facecolor= 'blue')
  plt.scatter(cluster_centers[0][0], cluster_centers[0][1],
            marker= "^", facecolor= 'red', s=300, edgecolors="k")
  plt.scatter(cluster_centers[1][0], cluster_centers[1][1],
            marker= "s", facecolor= 'green', s=300, edgecolors="k")
  plt.scatter(cluster_centers[2][0], cluster_centers[2][1],
            marker= "d", facecolor= 'blue', s=300, edgecolors="k")
```

```python
model = AffinityPropagation(preference=-50).fit(X)
```

```python
cluster_centers_indices = model.cluster_centers_indices_
labels = model.labels_
n_clusters = len(cluster_centers_indices)
print(cluster_centers_indices) #각 군집 대표의 인덱스를 도출한다 
print(labels)
```

```python
from itertools import cycle 

for k, col in zip(range(n_clusters), cycle("rgb")):
  cluster_members = labels == k 
  plt.plot(X[cluster_members, 0], X[cluster_members, 1], col+".")
  plt.plot(X[cluster_centers_indices[k], 0], X[cluster_centers_indices[k], 1], col+".", mec = 'k', mew=3, markersize=20)

  for i in X[cluster_members]:
    # plt.plot([i[0], X[cluster_centers_indices[k], 0]],
            #  [i[1], X[cluster_centers_indices[k],1]], col)


    plt.plot([i[0], X[cluster_centers_indices[k], 0]],
              [i[1], X[cluster_centers_indices[k], 1]], col)
  plt.plot(X[cluster_centers_indices[k], 0], X[cluster_centers_indices[k],1], col+".", mec= "k", mew=3 ,markersize = 20)
```
